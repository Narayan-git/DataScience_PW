```python
print('Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.')
print('ANSWERS: ')
''' 
Class:
    It's a blueprint or template that defines the properties (attributes) and behaviors (methods) that objects of a particular type will have.
    It's a logical abstraction, not a physical entity.
    It serves as a guideline for creating objects with consistent structure and functionality.

Object:
    It's a concrete instance of a class, created in memory.
    It has its own state (data stored in its attributes) and behavior (defined by its methods).
    It represents a specific entity with unique characteristics and actions.
    
'''
# Example
# Class:

class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age

    def bark(self):
        print("Woof!")

    def fetch(self):
        print("Fetching the ball!")

# Objects:

fido = Dog("Fido", "Labrador", 3)
buddy = Dog("Buddy", "Golden Retriever", 2)

fido.bark()  # Output: Woof!
buddy.fetch()  # Output: Fetching the ball!
```

```python
print('Q2. Name the four pillars of OOPs.')
print('ANSWERS: ')
print('''
    1. Abstraction: Simplifying complex objects by focusing on their essential features and behaviors, hiding internal details from the user.
    2. Encapsulation: Combining data and related functions into a single unit (object), controlling access to data and ensuring its integrity.
    3. Inheritance: Creating new classes (child classes) that inherit properties and behaviors from existing classes (parent classes), promoting code reuse and specialization.
    4. Polymorphism: Allowing objects to respond to the same message in different ways based on their type, increasing flexibility and code generality.''')
```

```python
print('Q3. Explain why the __init__() function is used. Give a suitable example.')
print('ANSWERS: ')
print('''
Purpose:
      -> The __init__() function, often called the constructor,
        is a special method in Python classes that is automatically invoked when a new object of that class is created.
      -> Its primary purpose is to initialize the attributes (data) of the object, ensuring that each object starts with a well-defined state.

Key Points:
      -> Automatic Execution: When you create an object using object_name = Class(), the __init__() method runs behind the scenes to set up the object's initial state.
      -> Attribute Assignment: Within the __init__() method, you typically assign values to the object's attributes using the self keyword, which refers to the newly created object itself.
      -> Customization: The __init__() method allows you to customize how objects are created, ensuring they have the necessary data and are ready for use.
Example:      
''')
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def introduce(self):
        print(f"Hi, my name is {self.name} and I'm in grade {self.grade}.")

# Create a new student object
student1 = Student("Alice", 10, "5th")

# Access and print student information
print(f"Student name: {student1.name}")
print(f"Student age: {student1.age}")
student1.introduce()  # Output: Hi, my name is Alice and I'm in grade 5th.
```

```python
print('Q4. Why self is used in OOPs?')
print('ANSWERS: ')
print('''
In OOPs, self is like a magic pointer.
It connects methods to the specific object they're called on, letting them access and manipulate its unique data and properties.
It ensures methods act on individual objects, not general functions, making code dynamic and flexible.
Think of it as the "this" object for each instance, always pointing to the right place for smooth operation.
''')
```

```python
print('Q5. What is inheritance? Give an example for each type of inheritance.')
print('ANSWERS: ')
print('''
nheritance is a core concept in object-oriented programming (OOP) that allows new classes to be created based on existing ones, inheriting their properties and methods. It promotes code reuse, reduces redundancy, and creates hierarchical relationships between classes.

Here are the main types of inheritance with examples:

1. Single Inheritance:

A child class inherits from a single parent class.
Example:

class Animal:
    # ... animal methods

class Dog(Animal):
    # ... dog-specific methods
      
2. Multiple Inheritance:

A child class inherits from multiple parent classes.
Example:
Python
class Musician:
    # ... musician methods

class Singer(Musician):
    # ... singer-specific methods

class Guitarist(Musician):
    # ... guitarist-specific methods

class SingerGuitarist(Singer, Guitarist):
    # ... combined methods

3. Multilevel Inheritance:

A child class inherits from a parent class, which itself inherits from another parent class.
Example:
Python
class Vehicle:
    # ... vehicle methods

class Car(Vehicle):
    # ... car-specific methods

class ElectricCar(Car):
    # ... electric car-specific methods


4. Hierarchical Inheritance:

Multiple child classes inherit from a single parent class.
Example:
Python
class Employee:
    # ... employee methods

class Manager(Employee):
    # ... manager-specific methods

class Developer(Employee):
    # ... developer-specific methods

5. Hybrid Inheritance:

A combination of multiple inheritance types.
Example:
Python
class Shape:
    # ... shape methods

class Circle(Shape):
    # ... circle-specific methods

class Rectangle(Shape):
    # ... rectangle-specific methods

class Ellipse(Circle, Rectangle):  # Hybrid inheritance
    # ... ellipse-specific methods

''')
```
